Функция ниже просто считает вероятности перехода для каждого начального состояния (перемножая строку на столбец):

```python
def count_probabilities(cls: np.array, prob_matrix: np.array, steps: int) -> None:
    """
    Count probabilities depending on initial state using Markov chain
    :param cls: numbers of classes
    :param prob_matrix:
    :param steps: number of steps
    """

    result = {}
    i = 0
    for c in cls:
        result[c] = 0
        # take ith state with prob equal to 1
        initial_prob = np.zeros(len(cls))
        initial_prob[i] = 1
        i += 1

        # walk given amount of steps for each state
        q_n = initial_prob
        print("Initial probabilities: ", initial_prob)
        for step in range(steps):
            q_next = np.matmul(q_n, prob_matrix)
            q_n = q_next.copy()

        print(f'Probabilities after {steps} steps stating at state {c}')
        print(q_n)
```

# Симуляция цепи Маркова

Функция simulate_chain выполняет симуляцию цепи для каждого начального состояния и выводит счетчик состояний в словаре (для каждого начального состояния). Если отнормировать по числу шагов, то получится процент времени нахождения ЦМ в каждом из состояний.
Функция принимает массив из номеров классов ($\text{E}_1$ или $\text{E}_2$ в нашем случае).
Матрицу вероятностей перехода (для каждого класса, $P_{\text{E}_1}$ и $ P_{\text{E}_2} $ соответственно; количество шагов в цепи и булево значение, которое контролирует вывод, для большого количества шагов вывод лучше отключать).

```python
def simulate_chain(cls: np.array, prob_matrix: np.array,
                   steps: int, to_print: bool) -> Dict[int, Dict[int, int]]:
    """
    Simulate Markov chain
    :param cls: numbers of classes
    :param prob_matrix:
    :param steps: number of steps
    :returns: dict, where key is the initial state and value is dict with counts
    for each state
    """

    result_matrix = {}
    trajectory = []
    for c in cls:
        # c - is initial state, we simulate chain for each state
        result = {i: 0 for i in cls}
        ind = np.where(cls == c)[0][0]
        prob = prob_matrix[ind]
        trajectory.append(c)

        for i in range(steps - 1):
            # pick 1 state from array depending on prob weights
            new_state = random.choices(cls, weights=prob, k=1)[0]

            trajectory.append(new_state)
            # we will count an average time for each class
            result[new_state] += 1

            ind = np.where(cls == new_state)[0][0]
            prob = prob_matrix[ind]

        if to_print:
            print(f'{trajectory} for {steps} steps depending '
                  f'on initial state {c}')

        trajectory.clear()
        result_matrix[c] = result

    return result_matrix
```

Теперь сравним результаты с вектором финальных вероятностей.

Возьмем результат из симуляции из 1 млн. шагов. Для наглядности отнормируем результаты на число шагов. Для этого нужна следующая функция:

```python
def normalize_res(res_prob: Dict[int, Dict[int, int]], steps: int) -> Dict[int, Dict[int, float]]:
    res = {}

    for c in res_prob.keys():
        res[c] = {i: k/steps for i, k in res_prob[c].items()}
    return res
```

Возьмем среднее для каждого из состояний. Для этого напишем следующую функцию:

```python
def calculate_average_for_each_row(
        data: Dict[int, Dict[int, Union[int, float]]]
        ) -> Dict[int, float]:
    """
    Calculate average along y-axis in dict (for each value in inner dict)
    :param data: dict with dicts as values,
    function calculates an average for each key
    :returns: dict with averages
    """

    # initialize an empty dictionary to store the sums and counts for each row
    sums_and_counts = {}

    # iterate over each key in the outer dictionary
    for key in data:
        # iterate over each key in the inner dictionary
        for inner_key, value in data[key].items():
            # if this row hasn't been processed yet, initialize it
            if inner_key not in sums_and_counts:
                sums_and_counts[inner_key] = {'sum': 0, 'count': 0}

            # add the current value to the sum and increment the count
            sums_and_counts[inner_key]['sum'] += value
            if value != 0:
              sums_and_counts[inner_key]['count'] += 1

    # calculate the average for each row
    averages = {key: value['sum'] / value['count'] for key, value in sums_and_counts.items()}

    return averages
```

# Нарисуем график для вероятности каждого класса в зависимости от количества шагов в цепи
И сравним с полученными финальными вероятностями

```python
def run_multiple_chains( n: int
        ) -> List[Tuple[int, Dict[int, Dict[int, float]]]]:
    """
    Calculate average along y-axis in dict (for each value in inner dict)
    :param n: maximum number of steps to be taken
    :returns: list of tuples where first val is the number of steps,
    second value is the result for each step
    """
    n = max(n, 1000)

    data = []
    for chain_len in range(10, n, 100):
        res = simulate_chain(E, P, chain_len, False)
        res = normalize_res(res, chain_len)
        res = calculate_average_for_each_row(res)
        data.append((chain_len, res))

    return data
```

```python
data = run_multiple_chains(10_000)
```
